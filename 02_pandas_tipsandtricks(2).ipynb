# 11- Split dataframe into two subsets


```python
#Download dataset
import seaborn as sns 
import pandas as pd
df = sns.load_dataset('titanic')
df.head()
```

```python
#to split the data we have to check the length of dataset
len(df)

#Also you can view by (df.shape); it will show you both columns and rows (891, 15)
```

```python
#Now split the dataset
from random import random
boat_1= df.sample(frac=0.50, random_state=1) #We are spliting 50% of dataset (you can split how much you want)
boat_1.shape
```

```python
boat_2= df.drop(boat_1.index)
boat_2.shape   #here we just take remaining data from boat_1
```

# 13- Filtering a dataset

```python
#if you want to check that what kind of variable is in specific column in dataset
#e.g here we will check for sex column

df.sex.unique()
```

```python
#if you want to take data of only one variable in the column then;
#e.g here we will check for sex column (only female)

df[(df.sex=="female")]

#here you can see down that sex column have only female data now
```

```python
#You can check for different other columns and variables also
df.survived.unique()
df.pclass.unique()
df.embarked.unique()
df.fare.unique() #etc
```

```python
#You can check for specific variables in column also with their shape
df.embark_town.unique()
df[(df.embark_town=="Southampton")].shape
```

```python
#You can check for specific variables within two column
#like we checked above that how many people move from Southampton, but if you want to check that,
# how many females only move from there then:

df[(df.embark_town=="Southampton") & 
   (df.sex=='female')]

#see the output in embark_town and sex column
```

```python
#You can check for specific variables within two column
#like we checked above that how many females move from Southampton, but if you want to check that,
# how many females move from southampton and also Queenstown then, like you can check for two values also:

df[((df.embark_town=="Southampton")|
   (df.embark_town=="Queenstown")) &
   (df.sex=='female')]

#see the output in embark_town and sex column
```

```python
#Another way to check unique values in column
df[df.embark_town.isin(["Queenstown", "Southampton"])]
```

```python
#If you want to check any value in any column
# e.g in age column we want to check that how many people are of age more than 30

df[df.age > 30]

#check the output
#same if you want to check for less than 30 (df[df.age < 30]
# if you want to check the shape then just put like this (df[df.age > 30]).shape
```
